```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("preprocessed_data.csv")
```

```python
dataset
```

```python
# Number of rows and Columns in dataset
print("The Numbers of Row: ",dataset.shape[0])
print("The Numbers of Column: ",dataset.shape[1])
```

```python
#dataset shape find
dataset.shape
```

Univarite Analysis

```python
#univariate histogram for dataset--->some set of range repeated
import seaborn as sb
sb.distplot(dataset["ssc_p"],hist=True,kde=True,kde_kws={'color':'red'})
```

Bivarite Analysis -- Distribution Polt

```python
#comparing the two column in dataset using jointplot
sb.jointplot(x='ssc_p',y='salary',data=dataset)

plt.show()
```

```python
#comparing the two column in dataset using jointplot with hexa heating graph
sb.jointplot(x='ssc_p',y='salary',data=dataset,kind='hex')
plt.show()
```

```python
#comparing the two column in dataset using jointplot with kernal density estimation 
sb.jointplot(x='ssc_p',y='salary',data=dataset,kind='kde')
plt.show()
```

Pair Plot

```python
#pair plot
sb.pairplot(dataset,hue = 'gender',diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
```

Categorical PLot

```python
sb.stripplot(x = "ssc_b",y="ssc_p",data=dataset)
```

```python
sb.stripplot(x="gender",y="ssc_p",data=dataset,jitter=True)
plt.show()
```

```python
sb.swarmplot(x="gender",y="ssc_p",data=dataset)
plt.show()
```

Distribution Observation 

```python
sb.violinplot(x="gender",y="ssc_p",data=dataset)
plt.show()
```

```python
sb.violinplot(x="gender",y="ssc_p",data=dataset,hue="gender")
plt.show()
```

```python
sb.violinplot(x="specialisation",y="salary",data=dataset,hue="gender")
plt.show()
```
